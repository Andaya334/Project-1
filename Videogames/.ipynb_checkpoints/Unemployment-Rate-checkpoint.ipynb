```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
csv_path = "unemployment-rate.csv"
```

```python
# Read with Pandas
census_df=pd.read_csv(csv_path)
census_df.head()
```

```python
census_df=census_df.rename(columns={"DATE": "Month",
                                    "UNRATE": "Unemployment Rate"
                                   })
census_df.head()
```

```python
census_group_df=census_df.groupby(["Month","Unemployment Rate"])
census_group_df.head()
```

```python
census_df['Indicator'] = ["High" if x>= 5 else "Low" for x in census_df['Unemployment Rate']]
 
print(census_df)
```

```python
indicator=census_df["Indicator"].unique()
indicator
```

```python
total_recessions=census_df["Indicator"].value_counts()
total_recessions
```

```python
total_recessions.count()
```

```python
unemployment_rate_df=census_df['Unemployment Rate']
unemployment_rate_df.describe()
```

```python
x_axis = np.arange(len(census_df))
tick_locations = [value+0.4 for value in x_axis]
```

```python
plt.figure(figsize=(50,8))
plt.bar(x_axis, census_df["Unemployment Rate"], color='r', alpha=0.5, align="edge")
plt.xticks(tick_locations, census_df["Month"], rotation="vertical")
```
